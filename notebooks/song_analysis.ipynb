## Song Attribute Trends over time

```python
attribute = 'Genre'

# start_date = '0'
start_date = '2020-01-01' # Beginning of year
end_date = '2020-12-31' # End of year

song_details_df = song_details_df[start_date < song_details_df['Date']]

song_details_df = song_details_df[song_details_df['Date'] <= end_date]

song_details_df
```

```python
scored = song_details_df.assign(score = 1 / song_details_df['Rank'])
scored
```

```python
# TODO: decide if it's better to group by attribute then by Date, or by Date and then attribute

grouped_song_details = scored.groupby([attribute, 'Date'])
# grouped_song_details = scored.groupby(['Date', attribute])

type(grouped_song_details)
```

```python
grouped_song_details["score"].sum()
# grouped_song_details["score"].sum().sort_values(ascending=False)
```

```python
# Separate by attribute and sum by month
scored_by_attribute = grouped_song_details["score"].sum()
attribute_set = {attr for attr, date in scored_by_attribute.index}
attribute_set

# Dict mapping attribute -> df with summed scores over time
attr_score_over_time = {}

for attr in attribute_set:
    df = scored_by_attribute[attr]
    df.index = pd.to_datetime(df.index)
    attr_score_over_time[attr] = df.groupby(pd.Grouper(freq='M')).sum()

attr_score_over_time
```

### Average song length of genres

```python
def convert_time_to_seconds(song_length):
    minute, seconds = song_length.split(":")
    return int(minute)*60 + int(seconds)
```

```python
def convert_seconds_to_time(song_length):
    minute = str(int(song_length // 60))
    seconds = int(song_length % 60)
    return minute + ":" + "{:02d}".format(seconds)
```

```python
song_length_seconds = song_details_df["Length"].map(convert_time_to_seconds, na_action='ignore')
with_seconds = song_details_df.assign(length_seconds = song_length_seconds)
with_seconds
```

```python
with_seconds.groupby(['Genre'])["length_seconds"].mean().map(convert_seconds_to_time, na_action='ignore')
```

```python
# With time dimension
with_seconds.groupby(['Genre', 'Date'])["length_seconds"].mean().map(convert_seconds_to_time, na_action='ignore')
```
